```python
url='https://www.bls.gov/oes/current/msa_def.htm#T'
response=requests.get(url)
soup=bs(response.text, 'html.parser')
```

```python
lis=[]
soup.find_all('ul')[1024].h3.findNextSibling('ul').li
# soup.find_all('ul')[1026].li
```
